```python
import networkx as nx
```

```python
data = '/content/facebook_combined.txt'
graph = nx.Graph()
```

```python
with open(data, 'r') as f:
  for line in f:
    u, v = map(int, line.strip().split())
    graph.add_edge(u,v)
```

```python
def recommend_frnds(graph, user):
  if user not in graph:
    return f"user {user} not found"
  friends = set(graph.neighbors(user)) #set of users' friends
  potential_friends = {} #dict to store potential friend recommendations
  for friend in friends:
    for mutual in graph.neighbors(friend): #check for mutal friends
      if mutual != user and mutual not in friends:
        potential_friends[mutual] = potential_friends.get(mutual, 0) + 1 #adds +1 if mutual friend exisits, else 0
  recommendations = sorted(potential_friends.items(), key = lambda x: x[1]) #sort the potential friends w.r.t mutual frnds to help identify which mutual frnd is mostly repeated
  return recommendations
```

```python
demo_user = 5
print(recommend_frnds(graph, demo_user))
```
